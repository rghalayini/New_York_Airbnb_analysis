```python
#plot the data with matplotlib: what are the top 10 neighbourhoods with Airbnb listing?

#listings_df['neighbourhood_group'].value_counts().plot(kind='bar')

listings_df['neighbourhood_group'].value_counts().head(10).plot(kind='bar')
```

```python
#plot the data with seaborn

import seaborn as sns
sns.countplot(data=listings_df, x='neighbourhood_group')
```

```python
#to order it

ordered_neighbourhood=listings_df['neighbourhood_group'].value_counts().index
sns.countplot(data=listings_df, x='neighbourhood_group', order=ordered_neighbourhood)
```

```python
#what's the influence of neighbourhood group on room type

listings_df['room_type'].unique()
sns.countplot(data=listings_df, x='neighbourhood_group', order=ordered_neighbourhood, hue='room_type')
```

```python
#what is the distribution of flat prices?

sns.distplot(listings_df['price'])
```

```python
#How many flats are more expensive than 500$/night? Could we remove those outliers from the visualization? 

affordable_df = listings_df[listings_df['price']<=500]
```

```python
sns.distplot(affordable_df['price'])
```

```python
affordable_df.price.mean()
```

```python
#what is the distribution of flat prices based on the neighbourhood group?

sns.violinplot(data=affordable_df, x="neighbourhood_group", y="price")
```

```python
#to read them better, make the figure bigger

from matplotlib import pyplot as plt

plt.figure(figsize=(15,8))
sns.violinplot(data=affordable_df, x="neighbourhood_group", y="price")
```

```python
#can we plot the listing on a map?

affordable_df.plot(
    kind='scatter',
    x='longitude',
    y='latitude',
    c='price',
    cmap='inferno',
    colorbar=True,
    alpha=0.8,
    figsize=(12,8))


```

```python
#Adding NYC map underneath!

import urllib


i = urllib.request.urlopen('https://upload.wikimedia.org/wikipedia/commons/e/ec/Neighbourhoods_New_York_City_Map.PNG')
plt.imshow(plt.imread(i), zorder=0, extent=[-74.258, -73.7, 40.49, 40.92])
ax = plt.gca()
affordable_df.plot(
    ax=ax,
    zorder=1,
    kind='scatter',
    x='longitude',
    y='latitude',
    c='price',
    cmap='inferno',
    colorbar=True,
    alpha=0.8,
    figsize=(12,8))
```
